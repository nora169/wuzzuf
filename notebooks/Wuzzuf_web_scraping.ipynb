```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import csv
import math
```

```python
link='https://wuzzuf.net/search/jobs/?a=navbl&q=data%20science%20jobs&start=1'
soup=requests.get(link).text
soup=BeautifulSoup(soup,'html.parser')
n=int(soup.find(attrs={'class':'css-8neukt'}).text.split('of')[1].strip())
pages_n=math.ceil(n/15)
```

```python
jobs=[]
companies=[]
countries=[]
all_talents=[]    

for i in range (0,pages_n):
    link=f'https://wuzzuf.net/search/jobs/?a=navbl&q=data%20science%20jobs&start={i}'
    soup=requests.get(link).text
    soup=BeautifulSoup(soup,'html.parser')
    job_soups=soup.find_all('h2',attrs={'class':'css-m604qf'})
    for job_soup in  job_soups:
        jobs.append(job_soup.find('a').text)
    companies_soup=soup.find_all('div',attrs={'class':'css-d7j1kk'})
    for company_soup  in companies_soup:
        companies.append(company_soup.find('a').text[:-2])
    countries_soup=soup.find_all('span',attrs={'class':'css-5wys0k'})
    for country_soup in countries_soup:
        countries.append(country_soup.text)
    all_soups=soup.find_all(attrs={'class':'css-y4udm8'})
    for all_soup in all_soups:
        soups=all_soup.find_all('a',attrs={'class':'css-o171kl'})
        talents=[]
        for talent_soup in soups[0]:
            talents.append(talent_soup)
        for talent_soup in soups[1:]:
            if ' · 'in talent_soup.text:
                talents.append(talent_soup.text[3:])
            else:
                talents.append(talent_soup.text)
        soups=all_soup.find_all('a',attrs={'class':'css-5x9pm1'})
        for talent_soup in soups:
            if ' · 'in talent_soup.text:
                talents.append(talent_soup.text[3:])
            else:
                talents.append(talent_soup.text)
        all_talents.append(talents)
    
```

```python
len(jobs),len(companies),len(countries),len(all_talents)
```

```python
with open('Wuzzuf_Data_Science.csv','w',encoding='utf-8',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['jobs','companies','countries','all_talents'])
    for i in range(len(jobs)):
        writer.writerow([jobs[i],companies[i],countries[i],all_talents[i]])
df=pd.read_csv('Wuzzuf_Data_Science.csv')
df   
```
